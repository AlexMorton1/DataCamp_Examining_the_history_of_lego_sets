## The Data

You have been provided with two datasets to use. A summary and preview are provided below.

## lego_sets.csv

| Column     | Description              |
|------------|--------------------------|
| `"set_num"` | A code that is unique to each set in the dataset. This column is critical, and a missing value indicates the set is a duplicate or invalid! |
| `"name"` | The name of the set. |
| `"year"` | The date the set was released. |
| `"num_parts"` | The number of parts contained in the set. This column is not central to our analyses, so missing values are acceptable. |
| `"theme_name"` | The name of the sub-theme of the set. |
| `"parent_theme"` | The name of the parent theme the set belongs to. Matches the name column of the parent_themes csv file.
|

## parent_themes.csv

| Column     | Description              |
|------------|--------------------------|
| `"id"` | A code that is unique to every theme. |
| `"name"` | The name of the parent theme. |
| `"is_licensed"` | A Boolean column specifying whether the theme is a licensed theme. |

```python
# Import pandas, read and inspect the datasets
import pandas as pd

lego_sets = pd.read_csv('data/lego_sets.csv')
lego_sets.head()
```

```python
parent_themes = pd.read_csv('data/parent_themes.csv')
parent_themes.head()
```

finding the percentage of all licensed sets ever released that are star wars themed

```python
# subsetting the licensed themes
licensed_themes = parent_themes[parent_themes['is_licensed'] == True]
```

```python
# subsetting lego sets for only licenced_sets
licenced_sets = lego_sets[lego_sets['parent_theme'].isin(licensed_themes['name'])]

# Creating a list of licensed themes and their percentage of total themed sets
licenced_sets_percentages = licenced_sets['parent_theme'].value_counts(normalize=True) * 100

# Percentage of Themed Lego sets that are starwars 
the_force = int(licenced_sets_percentages['Star Wars'])

# display the percentage of licenced sets that are starwars themed
print(the_force)
```

finding the year that the highest number of lego star wars sets were released

```python
# subsetting all lego starwars sets
lego_SW_sets = lego_sets[lego_sets['parent_theme'] == 'Star Wars']

# create a list of sets per Year
lego_SW_sets['SetCount'] = lego_SW_sets['name']
lego_SW_SetsPerYear = lego_SW_sets.groupby('year')['SetCount'].count()

# sort by SetCount ascending
lego_SW_SetsPerYearSorted = lego_SW_SetsPerYear.sort_values(ascending = False)

# get the year with the highest SetCount
new_era = int(lego_SW_SetsPerYearSorted.head(1).index[0])

# display the highest year 
print(new_era)
```
